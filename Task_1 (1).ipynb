```python
inventory_dataset.fillna({'Werksgüte': 'Not Available',
                          'Bestellgütentext': 'Not Available',
                          'Cluster': 'Not Available',
                          'Si-Gehalt': 'Not Available',
                          'Mn-Gehalt': 'Not Available',
                          'P-Gehalt': 'Not Available',
                          'S-Gehalt': 'Not Available',
                          'Cr-Gehalt': 'Not Available',
                          'Ni-Gehalt': 'Not Available',
                          'Mo-Gehalt': 'Not Available',
                          'PRODUCT_TYPE': 'Not Available',
                          'MATERIAL_NAME': 'Not Available',
                          'MATERIAL_QUALITY_NORM': 'Not Available',
                          'DEFECT_NOTES': 'Not Available',
                          'INCO_TERM': 'Not Available',
                          'MATERIAL_NUMBER': inventory_dataset['MATERIAL_NUMBER'].mean(),
                          'MIN/MAX_BID_EUR_PER_TON': inventory_dataset['MIN/MAX_BID_EUR_PER_TON'].mean()},
                         inplace=True)

```

```python
inventory_dataset.fillna({
    'Thickness (mm)': inventory_dataset['Thickness (mm)'].mean(),
    'Width (mm)': inventory_dataset['Width (mm)'].mean(),
    'Länge': inventory_dataset['Länge'].mean(),
    'Weight (kg)': inventory_dataset['Weight (kg)'].mean(),
    'V-Gehalt': inventory_dataset['V-Gehalt'].mean(),
    'Cu-Gehalt': inventory_dataset['Cu-Gehalt'].mean(),
    'Nb-Gehalt': inventory_dataset['Nb-Gehalt'].mean(),
    'Ti-Gehalt': inventory_dataset['Ti-Gehalt'].mean(),
    'Al-Gehalt': inventory_dataset['Al-Gehalt'].mean(),
    'B-Gehalt': inventory_dataset['B-Gehalt'].mean(),
    'Streckgrenze': inventory_dataset['Streckgrenze'].mean(),
    'Zugfestigkeit': inventory_dataset['Zugfestigkeit'].mean(),
    'Dehnung': inventory_dataset['Dehnung'].mean(),
    'ORDER_ID': inventory_dataset['ORDER_ID'].mean(),
    'SITE': 'Not Available',
    'SURFACE_COATING': inventory_dataset['SURFACE_COATING'].mean(),
    'LENGTH_MM': inventory_dataset['LENGTH_MM'].mean(),
    'HEIGHT_MM': inventory_dataset['HEIGHT_MM'].mean(),
    'NUMBER_OF_COILS': inventory_dataset['NUMBER_OF_COILS'].mean(),
    'DELIVERY_EARLIEST': 'Not Available',
    'DELIVERY_LATEST': 'Not Available',
    'BUY_NOW_EUR_PER_TON': inventory_dataset['BUY_NOW_EUR_PER_TON'].mean(),
    'CO2_PER_TON_MAX_KG': inventory_dataset['CO2_PER_TON_MAX_KG'].mean(),
    'VALID_UNTIL': 'Not Available'
}, inplace=True)
```

```python
print("Columns in inventory_dataset:")
print(inventory_dataset.columns)

# Checking column names in inventory_dataset
```

```python
inventory_dataset.rename(columns={
    'Nenndicke NNN.NN mm mit Dezimalpunkt': 'Thickness (mm)',
    'Breite': 'Width (mm)',
    'Gewicht (kg)': 'Weight (kg)','NOMINAL_THICKNESS_MM': 'Thickness (mm)',
    'WIDTH_MM': 'Width (mm)',
    'MASS_MIN_KG': 'Weight (kg)'
}, inplace=True)

# renaming the columns as most of the names are in german in one sheet and second sheet contains names in english.
```

```python
print("Columns in inventory_dataset:")
print(inventory_dataset.columns)
```

```python
columns_to_drop = ['SURFACE_COATING', 'HEIGHT_MM', 'NUMBER_OF_COILS',
                   'DELIVERY_EARLIEST', 'DELIVERY_LATEST', 'CO2_PER_TON_MAX_KG']

inventory_dataset.drop(columns=columns_to_drop, inplace=True)

# Save the updated dataset
inventory_dataset.to_csv('cleaned_inventory_dataset.csv', index=False)
```
